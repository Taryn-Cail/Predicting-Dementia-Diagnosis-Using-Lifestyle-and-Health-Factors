##Altering Chronic Disease Column

manipulating chronic disease column in df2 to have three seperate columns for heart disease, diabetes and hypertension,ignoring any other conditions


```python
#going through the chronic health conditions column and seperating it
#ASSUMING that NAN values are 0
df2['Chronic_Diabetes'] = df2['Chronic_Health_Conditions'].str.contains('Diabetes', case=False, na=False).astype(int)

df2['Hypertension'] = df2['Chronic_Health_Conditions'].str.contains('Hypertension', case=False, na=False).astype(int)

df2['CardiovascularDisease'] = df2['Chronic_Health_Conditions'].str.contains('Heart', case=False, na=False).astype(int)
```

```python
# dropping health conditions column
df2 = df2.drop(columns=['Chronic_Health_Conditions'])
```

```python
#displaying new dataset 2
df2.head()
```

```python
# making sure both columns are the same
(df2['Diabetic'] == df2['Chronic_Diabetes']).value_counts()
```

```python
#merging the two columns
df2['Diabetic'] = df2[['Diabetic', 'Chronic_Diabetes']].max(axis=1)
```

```python
#dropping the chronic diabetes row
df2 = df2.drop(columns=['Chronic_Diabetes'])
```

```python
# Renaming diabetic column
df2.rename(columns={'Diabetic': 'Diabetes'}, inplace=True)
```

```python
#displaying new dataset 2
df2.head()
```

##Cleaning Diet column

```python
#identify unique values
df2['Nutrition_Diet'].unique()
```

I am going to classify the mediteranean diet as an 8 because its statistically seen as a very healthy diet.

The Balanced Diet will be given an 7.

And the Low-Carb diet will be given a 6.

This is to allow the diet quality and nutrition_diets to be merged.

This is subjective to debate and I am not a nutritionist, nor do I know the exact rating the other datset used to determine its 1-10 scale rating.

```python
# Changing Nutrition_Diet to Diet Quality marker
df2['Nutrition_Diet'] = df2['Nutrition_Diet'].replace({
    'Mediterranean Diet': 8,
    'Balanced Diet': 7,
    'Low-Carb Diet': 6
})
```

```python
# Renaming diet column
df2.rename(columns={'Nutrition_Diet': 'DietQuality'}, inplace=True)
```

##Current Info

```python
# print all the data frames to compare and contrast information
#displaying new dataset 1
df1.info()

#displaying new dataset 2
df2.info()
```

##Cleaning Smoking column

```python
#Getting possibilities for smoking status
df2['Smoking_Status'].unique()
```

```python
# Changing status to if someone has ever smoked in there life to binary output
df2['Smoking_Status'] = df2['Smoking_Status'].replace({
    'Current Smoker': 1,
    'Never Smoked': 0,
    'Former Smoker': 1
})
```

```python
# Renaming smoking column
df2.rename(columns={'Smoking_Status': 'Smoking'}, inplace=True)
```

##Cleaning gender column

```python
# Changing male and female column to 1 = female and 0 = male
df2['Gender'] = df2['Gender'].replace({
    'Female': 1,
    'Male': 0
})
```

##Cleaning sleep column

```python
df2['Sleep_Quality'].unique()
```

```python
df1['SleepQuality'].unique()
```

Changing poor sleep to be 4 because that corresponds to the lowest number in the df1, and good sleep to be 8 because that is the hightest number, roughly.

```python
# Changing sleep status to numeric output
df2['Sleep_Quality'] = df2['Sleep_Quality'].replace({
    'Poor': 4,
    'Good': 8
})
```

```python
# Renaming sleep column
df2.rename(columns={'Sleep_Quality': 'SleepQuality'}, inplace=True)
```

##Cleaning physical activity column

```python
#identifying unique values
df2['Physical_Activity'].unique()
```

```python
df1['PhysicalActivity'].unique()
```

I have decided to label the various activity levels as such.

```python
# Changing physical activity status to numeric output
df2['Physical_Activity'] = df2['Physical_Activity'].replace({
    'Sedentary': 1,
    'Mild Activity': 4,
    'Moderate Activity': 6,
    'Active': 9
})
```

```python
# Renaming physical activity column
df2.rename(columns={'Physical_Activity': 'PhysicalActivity'}, inplace=True)
```

##Cleaning Family History Column

```python
df2['Family_History'].unique()
```

```python
# Changing family history status to numeric output
df2['Family_History'] = df2['Family_History'].replace({
    'No': 0,
    'Yes': 1
})
```

```python
# Renaming family history column
df2.rename(columns={'Family_History': 'FamilyHistory'}, inplace=True)
```

##Cleaning Education column

```python
df2['Education_Level'].unique()
```

```python
df1['EducationLevel'].unique()
```

```python
# Changing physical activity status to numeric output
df2['Education_Level'] = df2['Education_Level'].replace({
    'No School': 0,
    'Primary School': 1,
    'Secondary School': 2,
    'Diploma/Degree': 3
})
```

```python
# Renaming Education column
df2.rename(columns={'Education_Level': 'EducationLevel'}, inplace=True)
```

##Cleaning Depression column

```python
df2['AlcoholLevel'].unique()
```

```python
df2['AlcoholLevel'].describe()
```

I am assuming the original scale of this alcohol is units per week being roughly 20 units max per week. I am going to reverse this so it matches with df1 and is easier for interpretation when modelling.

##Finalizing Cleaning

I am choosing not to remove any rows with columns that have 0 in them because there are many bianry values.

```python
#Converting floats to integers, fill any NAN values with 0
#df1
df1['AlcoholConsumption'] = df1['AlcoholConsumption'].fillna(0).astype(int)
df1['DietQuality'] = df1['DietQuality'].fillna(0).astype(int)
df1['PhysicalActivity'] = df1['PhysicalActivity'].fillna(0).astype(int)
df1['SleepQuality'] = df1['SleepQuality'].fillna(0).astype(int)
#df2
df2['AlcoholConsumption'] = df2['AlcoholConsumption'].fillna(0).astype(int)
```
